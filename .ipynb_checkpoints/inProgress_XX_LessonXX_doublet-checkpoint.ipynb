# Doublet

We just saw the example of a source-sink pair where the two singularities were separated by a certain distance. We now consider that the singularities have the same strength magnitude $\sigma$ and that the distance $l$ approaches zero while the strength magnitude increase such that the product $\sigma l$ remains constant. We obtain a flow pattern called a doublet.

The strength of the doublet is defined by $\kappa = \sigma l$.

In cylindrical coordinates $\left(r,\theta\right)$, the stream function is given by:

$$\psi\left(r,\theta\right) = -\frac{\kappa}{2\pi}\frac{\sin\theta}{r}$$

In Cartesian coordinates, the stream function becomes:

$$\psi\left(x,y\right) = -\frac{\kappa}{2\pi}\frac{y}{x^2+y^2}$$

and the velicity field induced by the doublet at the origin at the point $\left(x,y\right)$ is:

$$u\left(x,y\right) = \frac{\partial\psi}{\partial y} = -\frac{\kappa}{2\pi}\frac{x^2-y^2}{\left(x^2+y^2\right)^2}$$

$$v\left(x,y\right) = -\frac{\partial\psi}{\partial x} = -\frac{\kappa}{2\pi}\frac{2x}{\left(x^2+y^2\right)^2}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
from IPython.core.display import clear_output
```

```python
N = 50                            # Number of points in each direction
xStart,xEnd = -2.0,2.0            # x-direction boundaries
yStart,yEnd = -1.0,1.0            # y-direction boundaries
x = np.linspace(xStart,xEnd,N)    # x 1D-array
y = np.linspace(yStart,yEnd,N)    # y 1D-array
X,Y = np.meshgrid(x,y)            # generation of the mesh grid
```

```python
class Doublet:
    def __init__(self,strength,x,y):
        self.strength = strength
        self.x,self.y = x,y
        
    # get velocity field
    def velocity(self,X,Y):
        self.u = -self.strength/(2*pi)*\
                ((X-self.x)**2-(Y-self.y)**2)/((X-self.x)**2+(Y-self.y)**2)**2
        self.v = -self.strength/(2*pi)*\
                2*(X-self.x)*(Y-self.y)/((X-self.x)**2+(Y-self.y)**2)**2
            
    # get stream function
    def streamFunction(self,X,Y):
        self.psi = -self.strength/(2*pi)*(Y-self.y)/((X-self.x)**2+(Y-self.y)**2)
```

```python
strengthDoublet = 1.0                                # source strength
xDoublet,yDoublet = 0.0,0.0                          # location of the source
doublet = Doublet(strengthDoublet,xDoublet,yDoublet) # instance of the class Singularity
doublet.velocity(X,Y)                                # calculate the velocity field
doublet.streamFunction(X,Y)                          # calculate the stream function
```

```python
# Plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,doublet.u,doublet.v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xDoublet,yDoublet,c='r',s=80,marker='o')
```

## Uniform flow past a doublet

```python
class Freestream:
    def __init__(self,Uinf,alpha):
        self.Uinf = Uinf
        self.alpha = alpha
        
    # get the velocity
    def velocity(self):
        self.u = Uinf*cos(self.alpha*pi/180)
        self.v = Uinf*sin(self.alpha*pi/180)
        
    # get stream function
    def streamFunction(self,X,Y):
        self.psi = + self.Uinf*cos(self.alpha*pi/180)*Y\
                   - self.Uinf*sin(self.alpha*pi/180)*X
```

```python
Uinf = 1.0                            # freestream speed
alpha = 0.0                           # angle of attack (deg)
freestream = Freestream(Uinf,alpha)   # instance of the class Freestream
freestream.velocity()                 # calculate the velocity field
freestream.streamFunction(X,Y)        # calculate the stream function
```

```python
# Superimposition of the doublet on the freestream flow
u = freestream.u + doublet.u
v = freestream.v + doublet.v
psi = freestream.psi + doublet.psi

# Plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.contour(X,Y,psi,levels=[0.0],colors='g',linewidths=2,linestyles='solid')
plt.scatter(xDoublet,yDoublet,c='r',s=80,marker='o')
```

```python
# Compute the pressure coefficient
Cp = 1.0-(u**2+v**2)/Uinf**2

# Plotting
size = 10
plt.figure(num=0,figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.contourf(X,Y,Cp,levels=np.linspace(-3.0,1.0,100))
plt.scatter(xDoublet,yDoublet,c='r',s=80,marker='o')
plt.contour(X,Y,psi,\
            levels=[0.0],\
            colors='g',linewidths=2,linestyles='solid')
```
