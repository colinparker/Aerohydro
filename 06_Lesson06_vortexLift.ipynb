# Lift generated by a cylinder

In one of the previous notebooks, we saw that a doublet can be added to a uniform flow to represent the streamlines around a particular shape: a cylinder. Looking at the pressure coefficient, we realized that the lift and drag induced by the flow on the cylinder was exactly zeros.

Question: is a symmetrical shape, such as a cylinder, able to generate lift? What if the cylinder is rotating? Have you heard about the Magnus effect?

In this lesson, we will add a doublet and a vortex to a uniform horizontal flow.


Let's start by importing libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
N = 50                            # Number of points in each direction
xStart,xEnd = -2.0,2.0            # x-direction boundaries
yStart,yEnd = -1.0,1.0            # y-direction boundaries
x = np.linspace(xStart,xEnd,N)    # x 1D-array
y = np.linspace(yStart,yEnd,N)    # y 1D-array
X,Y = np.meshgrid(x,y)            # generation of the mesh grid
```

First, we recall the equations we derived in the lesson about doublet. In Cartesian coordinates, a doublet located at the origin has a stream function given by

$$\psi\left(x,y\right) = -\frac{\kappa}{2\pi}\frac{y}{x^2+y^2}$$

from which we can derive the velocity components

$$u\left(x,y\right) = \frac{\partial\psi}{\partial y} = -\frac{\kappa}{2\pi}\frac{x^2-y^2}{\left(x^2+y^2\right)^2}$$

$$v\left(x,y\right) = -\frac{\partial\psi}{\partial x} = -\frac{\kappa}{2\pi}\frac{2xy}{\left(x^2+y^2\right)^2}$$

We consider a doublet of strength $\kappa=1.0$ located at the origin. Let's be lazzy and copy-paste the work we have already done on the doublet!

```python
kappa = 1.0                    # strength of the doublet
xDoublet,yDoublet = 0.0,0.0    # location of the doublet
```

```python
# function to compute the velocity components of a doublet
def getVelocityDoublet(strength,xd,yd,X,Y):
    u = - strength/(2*pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2*pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    return u,v

# function to compute the stream-function of a doublet
def getStreamFunctionDoublet(strength,xd,yd,X,Y):
    psi = - strength/(2*pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    return psi
```

```python
# computing the velocity components on the mesh grid
uDoublet,vDoublet = getVelocityDoublet(kappa,xDoublet,yDoublet,X,Y)

# computing the stream-function on the mesh grid
psiDoublet = getStreamFunctionDoublet(kappa,xDoublet,yDoublet,X,Y)
```

```python
Uinf = 1.0        # freestream speed
```

```python
# freestream velocity components
uFreestream = Uinf*np.ones((N,N),dtype=float)
vFreestream = np.zeros((N,N),dtype=float)

# stream-function
psiFreestream = Uinf*Y
```

```python
# superimposition of the doublet on the freestream flow
u = uFreestream + uDoublet
v = vFreestream + vDoublet
psi = psiFreestream + psiDoublet

# plotting
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xDoublet,yDoublet,c='#CD2305',s=80,marker='o')

# cylinder radius
R = sqrt(kappa/(2*pi*Uinf))
circle = plt.Circle((0,0),radius=R,color='#CD2305',alpha=0.5)
plt.gca().add_patch(circle)

# stagnation points
xStagn1,yStagn1 = +sqrt(kappa/(2*pi*Uinf)),0
xStagn2,yStagn2 = -sqrt(kappa/(2*pi*Uinf)),0
plt.scatter([xStagn1,xStagn2],[yStagn1,yStagn2],c='g',s=80,marker='o');
```

We now consider a vortex located at the origin with a positive strength $\Gamma$.In Cartesian coordinates, the stream-function is given by

$$\psi\left(x,y\right) = \frac{\Gamma}{4\pi}\ln\left(x^2+y^2\right)$$

and the velocity components by

$$u\left(x,y\right) = \frac{\Gamma}{2\pi}\frac{y}{x^2+y^2}$$

$$v\left(x,y\right) = -\frac{\Gamma}{2\pi}\frac{x}{x^2+y^2}$$

```python
gamma = 4.0                  # strength of the vortex
xVortex,yVortex = 0.0,0.0    # location of the vortex
```

Based on the equations above, we define two functions `getVelocityVortex()` and `getStreamFunctionVortex()` to compute the velocity components and the stream-function on our Cartesian grid given the strength and the location of the vortex.

```python
# function to compute the velocity components of a vortex
def getVelocityVortex(strength,xv,yv,X,Y):
    u = + strength/(2*pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u,v
```

```python
# function to compute the stream-function of a vortex
def getStreamFunctionVortex(strength,xv,yv,X,Y):
    psi = strength/(4*pi)*np.log((X-xv)**2+(Y-yv)**2)
    return psi
```

```python
# computing the velocity components on the mesh grid
uVortex,vVortex = getVelocityVortex(gamma,xVortex,yVortex,X,Y)

# computing the stream-function on the mesh grid
psiVortex = getStreamFunctionVortex(gamma,xVortex,yVortex,X,Y)
```

Now that we have all the necessary ingredients (uniform flow, doublet and vortex), we apply the principle of superposition.

```python
# superimposition of the doublet and the vortex on the freestream flow
u = uFreestream + uDoublet +uVortex
v = vFreestream + vDoublet +vVortex
psi = psiFreestream + psiDoublet +psiVortex
```

##### Challenge task

The challenge task of the previous notebook aims to calculate the radius of the cylinder created by the doublet in a uniform flow

$$R = \sqrt{\frac{\kappa}{2\pi U_\infty}}$$

The new challenge is to find where are located the stagnation points.


```python
# cylinder radius
R = sqrt(kappa/(2*pi*Uinf))

# stagnation points
xStagn1,yStagn1 = +sqrt(R**2-(gamma/(4*pi*Uinf))**2),-gamma/(4*pi*Uinf)
xStagn2,yStagn2 = -sqrt(R**2-(gamma/(4*pi*Uinf))**2),-gamma/(4*pi*Uinf)

# plotting
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
circle = plt.Circle((0,0),radius=R,color='#CD2305',alpha=0.5)
plt.gca().add_patch(circle)
plt.scatter(xVortex,yVortex,c='#CD2305',s=80,marker='o')
plt.scatter([xStagn1,xStagn2],[yStagn1,yStagn2],c='g',s=80,marker='o');
```

## Pressure coefficient

Now, we would like to plot the pressure coefficient on the surface of the cylinder and compare with the case with no vortex.

The velocity components in polar coordinates for the triplet {freestream,doublet,vortex} are given by

$$u_r\left(r,\theta\right) = U_\infty \cos\theta \left(1-\frac{R^2}{r^2}\right)$$

$$u_\theta\left(r,\theta\right) = -U_\infty \sin\theta \left(1-\frac{R^2}{r^2}\right) - \frac{\Gamma}{2\pi r}$$

where $R$ is the cylindar radius defined above.

From those equations, we see that the radial component vanishes on the surface of the cylinder whereas the tangential velocity is given by

$$u_\theta\left(R,\theta\right) = -2U_\infty \sin\theta - \frac{\Gamma}{2\pi R}$$.

As a note, when there is no vortex, the tangential velocity becomes

$$u_\theta\left(R,\theta\right) = -2U_\infty \sin\theta$$.

From the previous notebooks, we know that the pressure coefficient is defined by

$$C_p = 1-\frac{U^2}{U_\infty^2}$$

where $U^2 = u^2+v^2 = u_r^2+u_\theta^2$.

Let's plot the surface pressure coefficient!

```python
theta = np.linspace(0,2*pi,100)
utheta = -2*Uinf*np.sin(theta)-gamma/(2*pi*R)

Cp = 1 - (utheta/Uinf)**2

# if there was no vortex
utheta_noVortex = -2*Uinf*np.sin(theta)
Cp_noVortex = 1 - (utheta_noVortex/Uinf)**2

# plotting
size = 6
plt.figure(figsize=(size,size))
plt.grid(True)
plt.xlabel(r'$\theta$',fontsize=18)
plt.ylabel(r'$C_p$',fontsize=18)
plt.xlim(theta.min(),theta.max())
plt.plot(theta,Cp,color='#CD2305',linewidth=2,linestyle='-')
plt.plot(theta,Cp_noVortex,color='g',linewidth=2,linestyle='-')
plt.legend(['with vortex','without vortex'],loc='best',prop={'size':16});
```
