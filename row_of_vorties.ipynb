# Row of Vorticies

## Finite row

###### Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

###### Create mesh

```python
N = 50
xStart, xEnd = -2.0,2.0
yStart, yEnd = -0.5,0.5
x=np.linspace(xStart, xEnd, N)
y=np.linspace(yStart, yEnd, N)
X,Y=np.meshgrid(x,y)
```

###### Set Parameters

```python
gamma = 5.0 ##votex strength
nV=1000  ##number of vorticies
dxv=0.5 ##spacing

##initialize
yvortex = 0
ufield=0
vfield=0
psifield=0
```

###### Define velocity functions

```python
def getVelocityVortex(strength,xv,yv,X,Y):
    u = + strength/(2*pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u,v
    
def getStreamFunctionVortex(strength,xv,yv,X,Y):
    psi = strength/(4*pi)*np.log((X-xv)**2+(Y-yv)**2)
    return psi
```

###### Calulate Field

```python
for i in range(0,nV):
    xvortex=-(dxv*nV)*0.5+i*dxv
    uvortex,vvortex = getVelocityVortex(gamma,xvortex,yvortex,X,Y)
    psivortex=getStreamFunctionVortex(gamma,xvortex,yvortex,X,Y)
    ufield=ufield+uvortex
    vfield=vfield+vvortex
    psifield=psifield+psivortex
```

```python
Vstart=-dxv*nV*0.5
Vend=dxv*nV*0.5
XV=np.linspace(Vstart, Vend, nV)
YV=np.linspace(0, 0, nV)

size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title('Finite Row of Vorites',fontsize=20)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,ufield,vfield,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(XV,YV,c='red',s=10,marker='o');
```

## Infinite Row

###### Set spacing and find vorties position

```python
a=.5
Ninf=abs(xEnd-xStart)/a
gamma = 5.0
XVinf=np.linspace(xStart, xEnd, Ninf)
YVinf=np.linspace(0, 0, Ninf)
```

###### Solve velocity field

```python
uinfr=gamma/(2*a)*np.sinh(2*pi*Y/a)/(np.cosh(2*pi*Y/a)-np.cos(2*pi*X/a))
vinfr=gamma/(2*a)*np.sin(2*pi*X/a)/(np.cosh(2*pi*Y/a)-np.cos(2*pi*X/a))
```

###### Plot

```python
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title('Infinite Row of Vortices', fontsize=20)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uinfr,vinfr,\
         density=2,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(XVinf,YVinf,c='red',s=10,marker='o');
```

### By setting the spacing equal and increasing the number of vorties in the finite case you can see the two plots converge to similar solutions

```python
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.title('Finite',fontsize=20)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,ufield,vfield,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(XV,YV,c='red',s=10,marker='o');
```
