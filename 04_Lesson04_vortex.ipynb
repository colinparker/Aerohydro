# Vortex

So far, we have covered source an sink, shown that a source-sink pair in a uniform flow can used to model the flow over a oval, shown that a doublet in a uniform flow represents the streamlines around a cylinder. In this lesson, we will introduce a new type of singularity, named a vortex.

After importing our librairies and creating our evenly spaced Cartesian grid, the equations for the vortex flow will be derived.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
N = 50                            # Number of points in each direction
xStart,xEnd = -2.0,2.0            # x-direction boundaries
yStart,yEnd = -1.0,1.0            # y-direction boundaries
x = np.linspace(xStart,xEnd,N)    # x 1D-array
y = np.linspace(yStart,yEnd,N)    # y 1D-array
X,Y = np.meshgrid(x,y)            # generation of the mesh grid
```

Consider a flow pattern where all the streamlines are concentric circles about a given point. Moreover, the velocity, inversionally proportional to the distance to the central point is constant along a streamline. Such a flow is called a *vortex* and can be mathematically expessed in polar coordinates by

$$u_r\left(r,\theta\right) = 0 \quad \text{,} \quad u_\theta\left(r,\theta\right) = \frac{\text{constant}}{r}$$

The constant can be found by evaluating the circulation around a given circular streamline

$$\Gamma = -\oint_C \mathbf{V}\cdot d\mathbf{S} = -u_\theta 2 \pi r$$

Therefore

$$\text{contant} = \frac{\Gamma}{2\pi}$$

and

$$u_\theta\left(r,\theta\right) = \frac{\Gamma}{2\pi r}$$

Integrating the velocity components, we can get the stream-function

$$\psi\left(r,\theta\right) = \frac{\Gamma}{2\pi}\ln r$$

In Cartesian coordinates, the stream-function is given by

$$\psi\left(x,y\right) = \frac{\Gamma}{4\pi}\ln\left(x^2+y^2\right)$$

and the velocity components by

$$u\left(x,y\right) = \frac{\Gamma}{2\pi}\frac{y}{x^2+y^2}$$

$$v\left(x,y\right) = -\frac{\Gamma}{2\pi}\frac{x}{x^2+y^2}$$

We define a vortex of strength $\Gamma=5$ located at the origin.

```python
gamma = 5.0                  # strength of the vortex
xVortex,yVortex = 0.0,0.0    # location of the vortex
```

Based on the equations above, we define two functions `getVelocityVortex()` and `getStreamFunctionVortex()` to compute the velocity components and the stream-function on our Cartesian grid given the strength and the location of the vortex.

```python
# function to compute the velocity components of a vortex
def getVelocityVortex(strength,xv,yv,X,Y):
    u = + strength/(2*pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    return u,v
```

```python
# function to compute the stream-function of a vortex
def getStreamFunctionVortex(strength,xv,yv,X,Y):
    psi = strength/(4*pi)*np.log((X-xv)**2+(Y-yv)**2)
    return psi
```

```python
# computing the velocity components on the mesh grid
uVortex,vVortex = getVelocityVortex(gamma,xVortex,yVortex,X,Y)

# computing the stream-function on the mesh grid
psiVortex = getStreamFunctionVortex(gamma,xVortex,yVortex,X,Y)
```

Once the velocity components have been computed, we are now able to visualize the streamlines induced by a vortex: concentric circles around the vortex center as expected.

```python
# plotting
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,uVortex,vVortex,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xVortex,yVortex,c='#CD2305',s=80,marker='o');
```

## Vortex & Sink

The vortex defined on the first part can be now associated to other singularities, for instance an sink. As done in previous notebooks, we can code two functions to compute the velocity components and the stream-function induced by a sink (a remember that the sink can be easily replaced by a source by just changing the sign of the strength).

```python
strengthSink = -1.0         # strength of the sink
xSink,ySink = 0.0,0.0       # location of the sink
```

```python
# function to compute the velocity components of a sink
def getVelocitySink(strength,xs,ys,X,Y):
    u = strength/(2*pi)*(X-xs)/((X-xs)**2+(Y-ys)**2)
    v = strength/(2*pi)*(Y-ys)/((X-xs)**2+(Y-ys)**2)
    return u,v
```

```python
# function to compute the stream-function of a sink
def getStreamFunctionSink(strength,xs,ys,X,Y):
    psi = strength/(2*pi)*np.arctan2((Y-ys),(X-xs))
    return psi
```

```python
# computing the velocity components on the mesh grid
uSink,vSink = getVelocitySink(strengthSink,xSink,ySink,X,Y)

# computing the stream-function on the mesh grid
psiSink = getStreamFunctionSink(strengthSink,xSink,ySink,X,Y)
```

The visualize the streamlines induced by a vortex-sink pair, we apply the principle of superposition.

```python
# superimposition of the sink and the vortex
u = uVortex + uSink
v = vVortex + vSink
psi = psiVortex + psiSink

# plotting
size = 10
plt.figure(figsize=(size,(yEnd-yStart)/(xEnd-xStart)*size))
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.xlim(xStart,xEnd)
plt.ylim(yStart,yEnd)
plt.streamplot(X,Y,u,v,\
               density=2.0,linewidth=1,arrowsize=1,arrowstyle='->')
plt.scatter(xVortex,yVortex,c='#CD2305',s=80,marker='o');
```
